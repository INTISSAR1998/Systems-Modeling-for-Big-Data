**Application 1 :**

Nous allons construire une fonction signe, permettant de calculer le singe d'une valeur 
$$
\begin{cases}
1 & \text{ si } x > 0  \\ 
0 & \text{ si } x = 0 \\ 
-1 & \text{  sinon.} \\ 
\end{cases}
$$

```python
def signe(x):
    if x  == 0: 
        return 0 
    elif x > 0:
        return 1 
    else:
        return -1 
```

Exemple : Résoudre le système d'équations: :


Résoudre le système d'équations linéaires suivant : 
$$
(S) \begin{cases}
3 \ x + 6 \ y - 5 \ z & = 12 \\
x - 3 \ y + 2 \ z & = -2 \\
5\ x - y + 4 \ z & = 10 \\ 
\end{cases}
$$

```python
import numpy as np
a = np.array([[3,6,-5], [1,-3,2], [5,-1,4]])
b = np.array([12,-2,10])
x = np.linalg.solve(a, b)
x
```
